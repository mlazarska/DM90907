```python
class TestDM():  #definiujemy klasę, wynik dostaniemy dopiero po jej uzyciu
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    def predict(self,X_test):
        preditctions =[]
        for row in X_test:
            label = np.random.choice(self.y_train)
            preditctions.append(label)
        return preditctions 
```

```python
clfDM = TestDM()
clfDM.fit(X_tr, y_tr)
y_DM = clfDM.predict(X_te)
print(accuracy_score(y_te,y_DM))
#za każdym wywołaniem, dostaniemy różne wartości, bo za każdym razem tym razem losujemy wartość label
```

```python
#klasyfikator, który sprawdzi wszystkie odległości euklidesowe między elementem i resztą elementów i na tej podstwie zrobi
#nam klasę/y
from scipy.spatial import distance
def euc(a,b):
    return distance.euclidean(a,b)
```

```python
euc(5,2)
```

```python
class TestDM(): 
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    def predict(self,X_test):
        preditctions =[]
        for row in X_test:
            label = self.closest(row)
            preditctions.append(label)
        return preditctions 
    def closest(self,row):
        best_dist = euc(row,self.X_train[0])
        best_index = 0
        for i in range(1,len(self.X_train)):
            dist = euc(row,self.X_train[i])
            if dist < best_dist:
                best_dist = dist
                best_index = i
        return self.y_train[best_index]
```

```python
clfDM = TestDM()
clfDM.fit(X_tr, y_tr)
y_DM = clfDM.predict(X_te)
print(accuracy_score(y_te,y_DM))
```
