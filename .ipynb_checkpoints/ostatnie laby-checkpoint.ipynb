```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
df['target']=iris.target
```

```python
df=df[:100]
```

```python
test = asb(df,'target')
```

```python
print(test.DESCR)
```

```python
test.data
```

```python
test.fit()
```

```python
test.get_scorecard()
```

```python
test.model_info_
```

```python
test.fit(n_category=3)
#gdy bez parametru, to podzili wtedy intuicyjnie na 4 w linii 35
```

```python
test.show_stats("petal width (cm)")
```
